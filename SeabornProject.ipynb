```python
#import package
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load sample dataset

sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.scatterplot(x="tip",y="total_bill", data=tips, hue="time", size="size")
```

```python
sns.histplot(tips.tip,kde=True, bins=9)
plt.xticks(ticks=[i for i in range(0, int(tips.tip.max()) + 2)])
plt.ylabel("Amount")
plt.xlabel("Tip Value (in USD)")
plt.title("Histogram of Tip Values (in USD)")

plt.show()
```

## Conclusion: 
### Most Common Tips: 
The majority of tips are around 2-4 units, indicating that most customers give tips within this range.

### Distribution Width: 
The tip value distribution has a long right tail, meaning that while most tips are 2-4 units, there are also higher tips, but they are less frequent.




```python
sns.boxplot(x="sex",y="tip",data=tips)
plt.yticks(list(np.arange(0,10,1)))
plt.ylabel("Tip Value (in USD)")
plt.xlabel("")
plt.title("Gender Differences in Tipping ")

plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=tips, hue="sex")
plt.yticks(list(np.arange(0,10,1)))
plt.ylabel("Tip Value (in USD)")
plt.xlabel("Day of the Week")
plt.title("Tip Distribution by Day of the Week and Gender")
```

## Conclusion: 

### Biggest Tips:
Saturday shows the highest number of outliers for men and women, which means that there are few cases of very high tips.

### Median Tips : 
Median tips for women are higher than for men at Friday and Sunday.
Median tips for men are higher at Thursday and Saturday .



### Interquartile ranges (IQR):

The range between the first and third quartiles(IQR) for women is more consistent and narrower, suggesting more consistent tipping among women.
The IQR is wider for men on most days, indicating greater variability in tips.


```python
sns.stripplot(x="day",y="tip", data=tips, hue="sex", dodge=True)
plt.yticks(list(np.arange(0,10,1)))
plt.ylabel("Tip Value (in USD)")
plt.xlabel("Day of the Week")
plt.title("Tip Distribution by Day of the Week and Gender")
plt.show()
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", fill=True)
plt.ylabel("Total Bill (in USD)")
plt.yticks(list(np.arange(0,60,5)))
plt.xlabel("Tip Value (in USD)")
plt.grid()

plt.show()
```

### Conclusion:

Customers which spend around 10-20 USD were most likely to leave a tip between 2 and 3 USD.

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

### Conclusion:

The higher the bill, the higher the tip usually is. This is quite intuitive - customers who pay more usually tip more.
